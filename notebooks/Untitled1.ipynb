```python
import pandas as pd
import networkx as nx
from collections import Counter

# caminhos
solution_path = "../data/optimization/solution_simple.csv"

# escolha qual rede usar para pegar nome/partido
# pode ser a de covotação ou de coautoria, desde que os IDs batam com os da otimização
gexf_path = "../data/networks/covoting-2019_2020_2021_2022-20251208-051415.gexf"
# ou, se quiser usar a de coautoria:
# gexf_path = "../data/networks/coauthorship-network-2019_2020_2021_2022-20251208-051609.gexf"

# 1) ler solução
solution = pd.read_csv(solution_path)

# filtrar só os selecionados
selected_ids = solution.loc[solution["selected"] == 1, "id"].tolist()

# 2) ler a rede GEXF
G = nx.read_gexf(gexf_path)

# dependendo de como o grafo foi salvo, os nós podem ser strings
# então vamos trabalhar com strings
selected_ids_str = [str(i) for i in selected_ids]

print("Deputados selecionados:\n")

parties = []

for node_id in selected_ids_str:
    if node_id not in G.nodes:
        # se não encontrar, tenta sem cast, só por segurança
        if node_id.isdigit() and int(node_id) in G.nodes:
            attrs = G.nodes[int(node_id)]
        else:
            print(f"(AVISO) Deputado {node_id} não encontrado no grafo.")
            continue
    else:
        attrs = G.nodes[node_id]

    nome = attrs.get("label", "(sem nome)")
    partido = attrs.get("party", "(sem partido)")

    print(f"{node_id}: {nome} ({partido})")
    parties.append(partido)

# 3) contagem por partido
print("\nQuantidade de deputados por partido na coalizão:\n")

counter = Counter(parties)
for partido, qtd in sorted(counter.items()):
    print(f"{partido}: {qtd}")

```
